```python
import matplotlib.pyplot  as plt
import numpy as np

def gen_pair(b,delta,r=1e3,c=1e-6):
    idx_0 = np.arange(0,delta,0.0000001)
    idx_1 = np.arange(delta,0.005,0.0000001)
    val_0 = b/delta*(1-np.exp(-idx_0/r/c))
    val_1 = b/delta*(1-np.exp(-delta/r/c))*np.exp(-(idx_1-delta)/r/c)
    idx = list(idx_0)+list(idx_1)
    val = list(val_0)+list(val_1)
    print('\n')
    return [idx,val]
```

```python
import matplotlib
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize # import figsize
#figsize(12.5, 4) # 设置 figsize
plt.rcParams['savefig.dpi'] = 120 #图片像素
plt.rcParams['figure.dpi'] = 120 #分辨率
a = gen_pair(b=10**-4,delta=10**-3,r=10**3,c=10**-6)
b = gen_pair(b=10**-4,delta=5*10**-4,r=10**3,c=10**-6)
c = gen_pair(b=10**-4,delta=10**-4,r=10**3,c=10**-6)
d = gen_pair(b=10**-4,delta=10**-7,r=10**3,c=10**-6)
plt.plot(a[0],a[1])
plt.plot(b[0],b[1])
plt.plot(c[0],c[1])
plt.plot(d[0],d[1])
plt.show()
```
